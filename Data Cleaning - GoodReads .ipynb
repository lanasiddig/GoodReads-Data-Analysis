GoodReads Export of 11th of Nov, Data Cleaning. 

```python
#import the required libraries
import pandas as pd 
import numpy as np 
```

```python
#Import the data for the GoodReads Library Extracted on the 11th of November
df = pd.read_csv(r"C:\Users\lanas\OneDrive\Desktop\Data Sets\goodreads_library_export.csv")
df.head()
```

```python
#Missing data check 
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#dropping all the duplicates though I believe there will not be any. 
df.drop_duplicates()
df.head()
```

```python
#dropping the owned copies since the information included is irrelevant and inaccurate
df = df.drop(columns = "Owned Copies")
df = df.drop(columns = "Spoiler")
df = df.drop(columns = "Private Notes")
df
```

```python
#seperating the Title Column into 2 columns consisting of "Title" and "Series"
df [["Title", "Series"]] = df["Title"].str.split('(', expand = True)
df
```

```python
#removing all the all the Null Values from the dataframe 
df = df.fillna("")
df
```

```python
#removing the paraenthese at the end of the column series 
df ["Series"] = df["Series"].str.strip(')')
df
```

```python
#sort by author ascending order 
df.sort_values(by=['Author'],inplace=False,ascending=True)
```
