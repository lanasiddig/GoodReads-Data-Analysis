```python
# correlation (non-numerical to numerical values)
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized  
df_numerized.head() 
```

```python
#lets start looking at the correlation (works only in numerical fields)
#pearson, kendall, spearman correlation 
df.corr(method='pearson') #pearson aka default one
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlatin Matrix for Numerical Features')
plt.xlabel('Books Feature')
plt.ylabel('Books Features')
plt.show()
```
